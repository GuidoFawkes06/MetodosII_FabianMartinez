Considere la siguiente ecuación diferencial:
$$\frac{\mathrm{d} u}{\mathrm{d} t} = \alpha u , \space u(0) = u_0$$
Muestre que aplicando iterativamente se obtiene:
$$u_k = (1 + \alpha \Delta t)^k u_0$$
Para el caso donde  $\alpha <0$. Muestre que la solución numérica oscilará si $\Delta t > \frac {-1}{\alpha}$. Ajuste $\alpha = -1$ y muestre las soluciones oscilatorias para $\Delta t = 1.1, 1.5, 1.9 s$. Por otro lado, la solución exacta nunca oscila $u(t) = e^{\alpha t}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solucion_num(alfa,deltat,u0,tmax):
    
    t_=np.arange(0,tmax,deltat)
    
    u_=np.zeros_like(t_)
    
    for k in range(len(u_)):
        
        u_[k]=(1+ alfa*deltat)**k *u0
    
    return (t_,u_)
```

```python
def solucion_exac(t,alfa, u0):
    
    return u0*np.exp(alfa*t)
```

```python
alfa=-1
tmax=20
u0=1000
t=np.linspace(0,tmax,100)
```

```python
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot()
ax.plot(solucion_num(alfa,1.1,u0,tmax)[0],solucion_num(alfa,1.1,u0,tmax)[1],label="$\Delta T = 1.1 s$")
ax.plot(solucion_num(alfa,1.5,u0,tmax)[0],solucion_num(alfa,1.5,u0,tmax)[1],label="$\Delta T = 1.5 s$")
ax.plot(solucion_num(alfa,1.9,u0,tmax)[0],solucion_num(alfa,1.9,u0,tmax)[1],label="$\Delta T = 1.9 s$")
ax.plot(t,solucion_exac(t,alfa,u0),label="Sol.exacta")
ax.legend()
```
